```python
dataset['Height (ft)'] = dataset['Height (ft)'].replace('-',np.nan)
dataset['Height (ft)'] = dataset['Height (ft)'].astype(float)
dataset['Weight (lb)'] = dataset['Weight (lb)'].replace('-',np.nan)
dataset['Weight (lb)'] = dataset['Weight (lb)'].astype(float)
dataset['Intelligence'] = dataset['Intelligence'].replace('-',np.nan)
dataset['Intelligence'] = dataset['Intelligence'].astype(float)
dataset['Strength'] = dataset['Strength'].replace('-',np.nan)
dataset['Strength'] = dataset['Strength'].astype(float)
dataset['Speed'] = dataset['Speed'].replace('-',np.nan)
dataset['Speed'] = dataset['Speed'].astype(float)
dataset['Power'] = dataset['Power'].replace('-',np.nan)
dataset['Power'] = dataset['Power'].astype(float)
dataset['Durability'] = dataset['Durability'].replace('-',np.nan)
dataset['Durability'] = dataset['Durability'].astype(float)
dataset['Combat'] = dataset['Combat'].replace('-',np.nan)
dataset['Combat'] = dataset['Combat'].astype(float)
```

```python
dataset.drop(['Height (cm)','Weight (kg)'], axis=1, inplace=True)
```

```python
dataset = dataset.drop_duplicates(subset=['Name'])
```

```python
dataset1 = dataset.drop(['Biography name', 'Place of birth', 'First appearance','Work occupation', 'Work base', 'Skin color'], axis=1)
dataset1.dtypes[1:20]
```

```python
obj_df = dataset1.select_dtypes(include=['object']).copy()
obj_df.head()
```

```python
obj_df.shape
```

```python
obj_df[obj_df.isnull().any(axis=1)]
```

```python
obj_df_new = pd.get_dummies(obj_df, columns=['Gender', "Eye color",'Race','Hair color','Publisher', 'Alignment'], prefix=['Gender', "Eye",'Race','Hair','Publisher', 'Alignment'])
obj_df_new.drop(['Gender_-','Eye_-', 'Race_-','Hair_-','Alignment_-'], axis=1, inplace=True)
obj_df_new.head()
```

```python
obj_df_new.shape
```

```python
dataset2 = dataset1.drop(['Gender', "Eye color",'Race','Hair color','Publisher', 'Alignment'], axis=1)
```

```python
dataset2.head()
```

```python
dataset3 = pd.merge(dataset2, obj_df_new, how="left", on="Name")
dataset3.head()
```

```python
dataset4 = dataset3.set_index('Name')
```

```python
dataset5 = dataset4.dropna()
```

```python
print ("Rows     : " ,dataset5.shape[0])
print ("Columns  : " ,dataset5.shape[1])
print ("\nFeatures : \n" ,dataset5.columns.tolist())
print ("\nMissing values :  ", dataset5.isnull().sum().values.sum())
print ("\nUnique values :  \n",dataset5.nunique())
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(dataset5, dataset5)
```

```python
cosine_sim.shape
```

```python
#indices = pd.Series(dataset.index, index=dataset['Name']).drop_duplicates()
#indices[:10]
dataset6 = dataset5.reset_index()
indices = pd.Series(dataset6.index, index=dataset6['Name']).drop_duplicates()
indices[:10]
```

```python
# Function that takes in supertino as input and outputs most similar superhero
def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the hero that matches the title
    idx = indices[title]

    # Get the pairwsie similarity scores of all heros with that hero
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the heros based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the most similar hero
    sim_scores = sim_scores[1:2]

    # Get the hero indices
    hero_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return dataset6['Name'].iloc[hero_indices]
```

```python
get_recommendations('Abin Sur')
```

```python
#for x in dataset6['Name']:
  #print(get_recommendations(x).to_numpy())
similar_hero = np.array([get_recommendations(x) for x in dataset6['Name']])
dataset6['similar_hero'] = similar_hero
dataset7 = dataset6[['Name','similar_hero']]
```

```python
dataset8 = pd.merge(dataset7, dataset, how="left", on="Name")
dataset8.head()
```
